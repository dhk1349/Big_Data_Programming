```python
import pandas as pd
import numpy as np
```

```python
path='/Users/donghoon/Desktop/KHU/3-2/Lecture/빅데이터프로그래밍/1115Assignment/'
file_name=path+'airbnb.csv'
df_air=pd.read_csv(file_name)
```

```python
def avg_cal(lst_input):
    my_sum=0
    cnt=0
    for i in lst_input:
        if i!=0:
            my_sum+=i
            cnt+=1
    return my_sum/cnt
```

```python
df_air2=pd.DataFrame({'neighborhood':df_air['neighborhood'],'reviews':df_air['reviews'], 'price':df_air['price'], 'overall_satisfaction':df_air['overall_satisfaction']})
df_air2=df_air2.fillna(0)
df_air2=df_air2.groupby('neighborhood')
df_air2_1=pd.DataFrame(index=df_air2.groups.keys())#make index by region
#print(df_air2.max())
df_air2_1['max of price']=df_air2.max()['price']
df_air2_1['min of price']=df_air2.min()['price']
df_air2_1['max of reviews']=df_air2.max()['reviews']
df_air2_1['min of reviews']=df_air2.min()['reviews']

result_lst=[]
for i ,j in df_air2:
    temp_lst=[]
    for k in j:
        if k=='neighborhood':
            temp=j[k].values[0]
        elif k=='reviews' or k=='price' or k=='overall_satisfaction':
            temp=avg_cal(j[k])
        temp_lst.append(temp)
    result_lst.append(temp_lst)
    
#group, avg reviews, avg price, avg, satisfaction 순서
input_lst=[[],[],[],[]]
for i in result_lst:
    input_lst[0].append(i[0])
    input_lst[1].append(i[1])
    input_lst[2].append(i[2])
    input_lst[3].append(i[3])
df_air2_1['avg of reviews']=input_lst[1]
df_air2_1['avg of price']=input_lst[2]
df_air2_1['avg of satisfaction']=input_lst[3]


print (df_air2_1)
```

```python
df_air2_1.to_csv("sorted_neighborhood_factors.csv")
```
