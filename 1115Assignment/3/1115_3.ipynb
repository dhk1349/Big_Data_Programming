```python
import pandas as pd
import numpy as np
```

```python
path='/Users/donghoon/Desktop/KHU/3-2/Lecture/빅데이터프로그래밍/1115Assignment/'
file_name=path+'airbnb.csv'
df_air=pd.read_csv(file_name)
```

```python
def classification(number):
    if 0<=number and number<=100:
        return "0-100"
    elif 100<number and number<=200:
        return "100-200"
    elif 200<number and number<=300:
        return "200-300"
    elif 300<number and number<=400:
        return "300-400"
    elif 400<number and number<=500:
        return "400-500"
    elif 500<number and number<=1000:
        return "500-1000"
    elif 1000<number and number<=5000:
        return "1000-5000"
```

```python
def cal_avg(lst_input):
    my_sum=0
    cnt=0
    for i in lst_input:
        if i!=0:
            my_sum+=i
            cnt+=1
    if cnt==0:
        cnt=1
    return my_sum/cnt
```

```python
def cal_med(lst):
    #print(lst)
    """
    for i in lst:
        if i!=0:
            temp_lst.append(i)
    """
    lst_len=len(lst)
    lst.sort()
    if lst_len%2==0:#짝수개
        return (lst[int(lst_len/2)]+lst[int(lst_len/2-1)])/2
    else:
        return (lst[int((lst_len-1)/2)])
```

```python
def make_neighborhood_list(lst):
    temp_set={}
    for i in lst:
        temp_set.add(i)
    return temp_set
```

```python
price_index=[]
for i in list(df_air['price']):
    price_index.append(classification(i))
#print (price_index)
df_air['price_index']=price_index
df_air=df_air.fillna(0)
df_air_class=df_air.groupby('price_index')
for i, j in df_air_class:
    print(i,j)
```

```python
result_index=["0-100", "100-200", "200-300", "300-400","400-500", "500-1000","1000-5000"]
#
df_result=pd.DataFrame(index=result_index)
df_result
```

```python
#avg accomo, med accomo, avg bedrooms, med bedrooms, avg reviews, med reviews, neighborhood list, length
#index, neighbor, reviews, accommodates, bedrooms 순으로 반복
result_lst=[]
for i,j in df_air_class:
    temp_lst=[]
    #나오는 index 순서가 섞임
    #i는 index 이름 j는 테이블 k는 column명
    temp_lst.append(i)
    for k in j: 
        if k=="accommodates" or k=="bedrooms" or k=="reviews":
            temp_lst.append(list(j[k]))
        elif k=="neighborhood":
            temp_lst.append(list(j[k]))
        else:
            group_len=(len(list(j[k])))
    temp_lst.append(group_len)
    result_lst.append(temp_lst)
print(result_lst)
```

```python
#index, neighbor, reviews, accommodates, bedrooms, length 순으로 반복
#이걸 column 별로 재정렬
accommodates_avg={}
accommodates_med={}
bedrooms_avg={}
bedrooms_med={}
reviews_avg={}
reviews_med={}
neighbor_list={}
length={}
for i in result_lst:
    #i는 index 하나의 정보
    #print(i[0])
    #print(i[3])
    accommodates_avg[i[0]]=cal_avg(i[3])
    accommodates_med[i[0]]=cal_med(i[3])
    bedrooms_avg[i[0]]=cal_avg(i[4])
    bedrooms_med[i[0]]=cal_med(i[4])
    reviews_avg[i[0]]=cal_avg(i[2])
    reviews_med[i[0]]=cal_med(i[2])
    temp_set=set()
    for j in i[1]:
        temp_set.add(j)
    neighbor_list[i[0]]=temp_set
    length[i[0]]=i[5]
```

```python
print (accommodates_avg,
accommodates_med,
bedrooms_avg,
bedrooms_med,
reviews_avg,
reviews_med,
neighbor_list,
length)
```

```python
def sort_and_push(colname, dictionary, DataFrame):
    temp_lst=[]
    temp_lst.append(dictionary['0-100'])
    temp_lst.append(dictionary['100-200'])
    temp_lst.append(dictionary['200-300'])
    temp_lst.append(dictionary['300-400'])
    temp_lst.append(dictionary['400-500'])
    temp_lst.append(dictionary['500-1000'])
    temp_lst.append(dictionary['1000-5000'])
    DataFrame[str(colname)]=temp_lst
```

```python
sort_and_push("accommodates_avg", accommodates_avg, df_result)
sort_and_push("accommodates_med", accommodates_med, df_result)
sort_and_push("bedrooms_avg", bedrooms_avg, df_result)
sort_and_push("bedrooms_med", bedrooms_med, df_result)
sort_and_push("reviews_avg", reviews_avg, df_result)
sort_and_push("reviews_med", reviews_med, df_result)
sort_and_push("neighbor_list", neighbor_list, df_result)
sort_and_push("length", length, df_result)
df_result
```
