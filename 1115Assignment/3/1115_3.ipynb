```python
def classification(number):
    if 0<=number and number<=100:
        return "0-100"
    elif 100<number and number<=200:
        return "100-200"
    elif 200<number and number<=300:
        return "200-300"
    elif 300<number and number<=400:
        return "300-400"
    elif 400<number and number<=500:
        return "400-500"
    elif 500<number and number<=1000:
        return "500-1000"
    elif 1000<number and number<=5000:
        return "1000-5000"
```

```python
def cal_avg(lst_input):
    my_sum=0
    cnt=0
    for i in lst_input:
        if i!=0:
            my_sum+=i
            cnt+=1
    return my_sum/cnt
```

```python
def cal_med(lst):
    temp_lst=[]
    for i in lst:
        if i!=0:
            temp_lst.append(i)
    temp_lst=temp_lst.sort()
    lst_len=len(temp_lst)
    if lst_len%2==0:#짝수개
        return (temp_lst[lst_len/2]+temp_lst[lst_len/2-1])/2
    else:
        return (temp_lst[(lst_len-1)/2])
```

```python
def make_neighborhood_list(lst):
    temp_set={}
    for i in lst:
        temp_set.add(i)
    return temp_set
```

```python
price_index=[]
for i in list(df_air['price']):
    price_index.append(classification(i))
#print (price_index)
df_air['price_index']=price_index
df_air=df_air.fillna(0)
df_air_class=df_air.groupby('price_index')
for i, j in df_air_class:
    print(i,j)
```
